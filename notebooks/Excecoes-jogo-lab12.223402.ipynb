## Descrição das classes de exceção

Classe | Descrição
----- | -----
ConstrucaoInvalida | Engloba todos os comandos de construção feitos pelo jogados que não seriam aceitos.
PecaNaoSelecionada | Indica que nenhuma peça foi selecionada ou a peça selecionada não está disponível na fase.
RecursoInsuficiente | Indica que o custo de construção da peça selecionada é maior do que a quantidade de recurso do jogador.
PosicaoInvalida | Indica que o jogador selecionou uma posição ocupada para construir uma peça.




```java
public class ConstrucaoInvalida extends Exception {
   public ConstrucaoInvalida() {
      super();
   }

   public ConstrucaoInvalida(String message) {
      super(message);
   }
}
```

```java
public class PecaNaoSelecionada extends ConstrucaoInvalida {
   public PecaNaoSelecionada() {
      super();
   }

   public PecaNaoSelecionada(String message) {
      super(message);
   }
}
```

```java
public class RecursoInsuficiente extends ConstrucaoInvalida {
   public RecursoInsuficiente() {
      super();
   }

   public RecursoInsuficiente(String message) {
      super(message);
   }
}
```

```java
public class PosicaoInvalida extends ConstrucaoInvalida {
   public PosicaoInvalida() {
      super();
   }

   public PosicaoInvalida(String message) {
      super(message);
   }
}
```

```java
public class Construir {
   public static String constroe(int x, int y, int peca, int custo, int recurso)  throws ConstrucaoInvalida {
       //Este código de texte ilustrativo supondo que o tabuleiro do jogo é um tabuleiro 3x3 
       //Com as posiçoes (1,1), (2,2) e (3,3) ocupadas
       //As peças disponiveis estão numeradas de 1 a 6.
       //No projeto final a função construir deverá receber como parametro também um objeto da classe tabuleiroe checar as posições validas
       //E deverá fazer alterações no próprio tabuleiro
       if (peca > 6 || peca < 1){
           throw new PecaNaoSelecionada("Você não selecionou uma peça válida");
       }
       if (x==y || x > 3 || y > 3 || x < 1 || y < 1){
           throw new PosicaoInvalida("Esta posição não pode ser selecionada");
       }
       if (custo > recurso){
           throw new RecursoInsuficiente("Você não tem recursos o suficiente para esta construção");
       }
       String divisao = "tabuleiro atualizado";
       return divisao;
    }
}
```

```java
System.out.println("teste_1");

try {
    String teste = Construir.constroe(1,1,1,1,1);
    System.out.println(teste);
} catch (PecaNaoSelecionada erro){
    System.out.println(erro.getMessage());
} catch (PosicaoInvalida erro){
    System.out.println(erro.getMessage());
} catch (RecursoInsuficiente erro){
    System.out.println(erro.getMessage());
}

System.out.println("teste_2");

try {
    String teste = Construir.constroe(1,1,7,1,1);
    System.out.println(teste);
} catch (PecaNaoSelecionada erro){
    System.out.println(erro.getMessage());
} catch (PosicaoInvalida erro){
    System.out.println(erro.getMessage());
} catch (RecursoInsuficiente erro){
    System.out.println(erro.getMessage());
}

System.out.println("teste_3");

try {
    String teste = Construir.constroe(1,2,3,20,10);
    System.out.println(teste);
} catch (PecaNaoSelecionada erro){
    System.out.println(erro.getMessage());
} catch (PosicaoInvalida erro){
    System.out.println(erro.getMessage());
} catch (RecursoInsuficiente erro){
    System.out.println(erro.getMessage());
}

System.out.println("teste_4");

try {
    String teste = Construir.constroe(1,2,3,10,10);
    System.out.println(teste);
} catch (PecaNaoSelecionada erro){
    System.out.println(erro.getMessage());
} catch (PosicaoInvalida erro){
    System.out.println(erro.getMessage());
} catch (RecursoInsuficiente erro){
    System.out.println(erro.getMessage());
}
```
